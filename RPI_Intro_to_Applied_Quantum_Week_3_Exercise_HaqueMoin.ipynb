```python
# import packages for workshop calculations and visualization

import numpy as np
import random
from matplotlib import pyplot as plt
```

```python
# setup function to simulate Bernoulli trial using probability parameter (p)

def get_Bernoulli(p):
    x = random.uniform(0, 1)      # generate random number between 0 and 1
    if x < p:
        return 0
    else:
        return 1
```

```python
# test and visualize the Bernoulli simulation

Ns = 100;                         # number of trials
p = 1/3                           # probability parameter

observations = [get_Bernoulli(p) for i in range(Ns)]

plt.hist(observations)
plt.show()
```

```python
# function to calculate the Monte Carlo estimate of the probability parameter p) by running (Ns) Bernoulli trials

def compute_MCE(p, Ns):
    return [get_Bernoulli(p) for i in range(Ns)].count(0)/Ns
```

```python
# Experiment demonstrating (visually) how accuracy of the Monte Carlo estimator improves with sample size

# Fix the parameter of the experiment
p = 1/5
max_shots = 1000000
shots = np.linspace(100,max_shots, num = 100, dtype = int)

# Run the experiment
accuracy = []
for Ns in shots:
    p_approx = compute_MCE(p, Ns)
    std = np.sqrt(p_approx * (1-p_approx)/Ns)
    accuracy.append(std)

#Visualize the results
plt.plot(shots, accuracy, '-x', label = 'MCE')
plt.plot(shots, 1/np.sqrt(shots), 'r-+', label = '1/$\sqrt{Ns}$')

plt.legend()
plt.yscale("log")
plt.xscale("log")
plt.show()
```
